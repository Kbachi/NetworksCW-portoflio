```python
import random
print ("Hello, World!")

# Addition
def add_numbers(a, b):
    return a + b

result = add_numbers(10, 5)
print("Addition Result:", result)

# Subtraction
def subtract_numbers(a, b):
    return a - b

result = subtract_numbers(10, 5)
print("Subtraction Result:", result)

# Multiplication
def multiply_numbers(a, b):
    return a * b

result = multiply_numbers(10, 5)
print("Multiplication Result:", result)

# Division
def divide_numbers(a, b):
    if b != 0:
        return a / b
    else:
        return "Division by zero is not allowed"

result = divide_numbers(10, 5)
print("Division Result:", result)

# Remainder
def remainder_numbers(a, b):
    return a % b

result = remainder_numbers(10, 5)
print("Remainder Result:", result)

# Power
def power_numbers(a, b):
    return a ** b

result = power_numbers(10, 5)
print("Power Result:", result)

# Python Data Types

# Create a list of 5 different fruits and print the list
fruits = ["Apple", "Banana", "Cherry", "Date", "Elderberry"]
print("List of fruits:", fruits)

# Create a tuple of 3 different animals and print the tuple
animals = ("Lion", "Tiger", "Bear")
print("Tuple of animals:", animals)

# Create a dictionary of 3 different key-value pairs where the keys are names of programming languages and the values are their release dates
languages = {
    "Python": 1991,
    "Java": 1995,
    "JavaScript": 1995
}
print("Dictionary of programming languages:", languages)

# Create a set of 3 different colours and print the set
colours = {"Red", "Green", "Blue"}
print("Set of colours:", colours)

# Create a string variable that contains the sentence "Python is an easy-to-learn programming language."
sentence = "Python is an easy-to-learn programming language."
print("String sentence:", sentence)

# Script to check if a number is positive, negative, or zero
number = float(input("Enter a number: "))
if number > 0:
    print("The number is positive.")
elif number < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

# Script to check if a string starts with a vowel or consonant
string = input("Enter a string: ").lower()
if string[0] in 'aeiou':
    print("The string starts with a vowel.")
else:
    print("The string starts with a consonant.")

# Script to check if two numbers are equal or not
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
if num1 == num2:
    print("The numbers are equal.")
else:
    print("The numbers are not equal.")

# Loops

# Script to print the numbers from 1 to 10 using a for loop
for i in range(1, 11):
    print(i)

# Script to print the even numbers from 2 to 10 using a for loop
for i in range(2, 11, 2):
    print(i)

# Script to print the first 5 numbers of the Fibonacci sequence using a while loop
a, b = 0, 1
count = 0
while count < 5:
    print(a)
    a, b = b, a + b
    count += 1

# Functions

# Function that takes two numbers as parameters and returns their sum
def sum_numbers(a, b):
    return a + b

# Function that takes a list of numbers as a parameter and returns the largest number in the list
def largest_number(numbers):
    return max(numbers)

# Function that takes a string as a parameter and returns the number of vowels in the string
def count_vowels(string):
    vowels = 'aeiou'
    return sum(1 for char in string.lower() if char in vowels)
# Password strength checker
def check_password_strength(password):
    special_characters = "!@#$%^&*()-+"
    if len(password) < 8:
        return "Password must be at least 8 characters long."
    if not any(char.isupper() for char in password):
        return "Password must contain at least one uppercase letter."
    if not any(char.islower() for char in password):
        return "Password must contain at least one lowercase letter."
    if not any(char.isdigit() for char in password):
        return "Password must contain at least one digit."
    if not any(char in special_characters for char in password):
        return "Password must contain at least one special character."
    return "Password is strong."

# Example usage
password = input("Enter a password: ")
print(check_password_strength(password))

# Caesar Cipher

def caesar_cipher(message, key, mode):
    result = ""
    for char in message:
        if char.isalpha():
            shift = key if mode == "encrypt" else -key
            shifted = ord(char) + shift
            if char.isupper():
                result += chr((shifted - 65) % 26 + 65)
            else:
                result += chr((shifted - 97) % 26 + 97)
        else:
            result += char
    return result

# Example usage
message = input("Enter a message: ")
key = int(input("Enter a key: "))
mode = input("Enter mode (encrypt/decrypt): ").lower()
print(caesar_cipher(message, key, mode))


def hangman_game(word_list, max_guesses):
    word = random.choice(word_list)
    guessed_word = ["_"] * len(word)
    guessed_letters = set()
    incorrect_guesses = 0

    print("Welcome to Hangman!")
    print(" ".join(guessed_word))

    while incorrect_guesses < max_guesses and "_" in guessed_word:
        guess = input("Guess a letter: ").lower()
        
        if guess in guessed_letters:
            print("You already guessed that letter. Try again.")
            continue
        
        guessed_letters.add(guess)
        
        if guess in word:
            for i, letter in enumerate(word):
                if letter == guess:
                    guessed_word[i] = guess
            print("Correct guess!")
        else:
            incorrect_guesses += 1
            print(f"Incorrect guess! You have {max_guesses - incorrect_guesses} guesses left.")
        
        print(" ".join(guessed_word))
    
    if "_" not in guessed_word:
        print(f"Congratulations, you guessed the word '{word}'!")
    else:
        print(f"Game over! The word was '{word}'.")

# Example usage
word_list = ["python", "java", "javascript"]
max_guesses = 6
hangman_game(word_list, max_guesses)
```
